```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import datetime as dt
from datetime import timedelta
```

```python
df_confirmed1 = pd.read_csv("time_series_covid19_confirmed_global.csv")
df_confirmed1.drop(columns=[ 'Province/State','Lat', 'Long', 'Country/Region'], inplace=True)
df_confirmed1.head()
```

```python
df_confirmed1_t = df_confirmed1.T
```

```python
df_confirmed1_t.head()
```

```python
df_confirmed1_t['confirmed']=df_confirmed1_t.sum(axis=1)
```

```python
df_confirmed1_t.head()
```

```python
df_confirmed1_t.info()
```

```python
df_confirmed1_t.drop(df_confirmed1_t.iloc[:, 0:253], inplace = True, axis = 1) 
```

```python
df_confirmed1_t.head()
```

```python
plt.figure(figsize=(20,5))
plt.plot(df_confirmed1_t["confirmed"],label="Confirmed Cases")
plt.ylabel("Number of Patients")
plt.xlabel("Date")
plt.xticks(rotation=90)
plt.legend()
```

```python
df_recovered1 = pd.read_csv("time_series_covid19_recovered_global.csv")
df_recovered1.drop(columns=[ 'Province/State','Lat', 'Long', 'Country/Region'], inplace=True)
df_recovered1.head()
```

```python
df_recovered1_t = df_recovered1.T
```

```python
df_recovered1_t.head()
```

```python
df_recovered1_t['recovered']=df_recovered1_t.sum(axis=1)
df_recovered1_t.head()
```

```python
df_recovered1_t.drop(df_recovered1_t.iloc[:, 0:239], inplace = True, axis = 1) 
df_recovered1_t.head()
```

```python
plt.figure(figsize=(20,5))
plt.plot(df_recovered1_t["recovered"],label="recovered Cases")
plt.ylabel("Number of Patients recovered")
plt.xlabel("Date")
plt.xticks(rotation=90)
plt.legend()
```

```python
df_deaths1 = pd.read_csv("time_series_covid19_deaths_global.csv")
df_deaths1.drop(columns=[ 'Province/State','Lat', 'Long', 'Country/Region'], inplace=True)
df_deaths1.head()
```

```python
df_deaths1_t = df_deaths1.T
df_deaths1_t['deaths']=df_deaths1_t.sum(axis=1)
```

```python
df_deaths1_t.head()
```

```python
df_deaths1_t.drop(df_deaths1_t.iloc[:, 0:253], inplace = True, axis = 1) 
df_deaths1_t.head()
```

```python
plt.figure(figsize=(20,5))
plt.plot(df_deaths1_t["deaths"],label="death Cases")
plt.ylabel("Number of people dead")
plt.xlabel("Date")
plt.xticks(rotation=90)
plt.legend()
```

```python
plt.figure(figsize=(20,6))
plt.plot(df_confirmed1_t["confirmed"],label="Confirmed Cases")
plt.plot(df_recovered1_t["recovered"],label="Recovered Cases")
plt.plot(df_deaths1_t["deaths"],label="Death Cases")
plt.ylabel("Number of Patients")
plt.xlabel("Date")
plt.legend()
plt.title("Growth Rate Plot for different Types of cases in word")
plt.xticks(rotation=90)
plt.show() 
```
